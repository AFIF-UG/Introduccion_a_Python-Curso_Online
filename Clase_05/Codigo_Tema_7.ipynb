#**Condicionales**

Gracias a las estructuras de control, podemos cambiar el flujo de ejecución de un programa, haciendo que ciertos bloques de código se ejecuten si y solo si se dan unas condiciones particulares.

Para establecer un condicional, iniciamos con la palabra "if" que traducida al español significa "si". Por ejemplo, cuando se habla en condicional, se espera realizar una acción bajo una **CONDICIÓN**: **Si** quieres ver televisión, debes terminar tu tarea.

**ESTRUCTURA DEL IF**

if *Condición*:

>  *Instruccion si se cumple la condición*

else:

>  *Instrucción si **No** se cumple la condición*


La parte de *Condición* es un valor booleano, osea solo puede valer **True** o **False**.


```
# CÓDIGO QUE SIEMPRE SE EJECUTA

if True:                  # NUNCA olvides poner los dos puntos ":" al final de una condición.
  print("Hola")       # También debe haber sangría en la acción, para estar seguro donde empezar a escribir
                      # puedes mantener el cursor después de los dos puntos y dar enter. 
```

```
# Programa que siempre ejecutará el else

if False:
  print("Hola")
else:
  print("Adiós")
```

#**ELIF**

Si tu quisieras agregar más de una condición para realizar diferentes acciones, puedes usar **elif** para este caso.

```
Hambre = True
Sed = False

if not Hambre:                 # CON "not" cambias el valor booleano, not True = False y not False = True
  print("Come algo")
elif not Sed:
  print("Bebe algo")
else:
  print("Estás satisfecho")
```

# **OR** Y **AND**



Si a diferencia del anterior caso, tú quieres hacer una misma acción para varias condiciones, puedes usar el **or** o el **and**.

**or**

Una de las condiciones debe ser verdad para que el condicional sea True.

**and**

Todas las condiciones deben ser verdad para que el condicional sea True.

---

 **OJO Si quieres pasar de renglón y continuar la condición debes encerrar todo entre paréntesis "( )"**

```
Carne = "Doble triple"
Salsa = "De Coral"
Decorado = "De lujo"
Estilo = "Animal"
Guijarros = "Extra"
Vibración = "Exprimida"
Eje = "Con grasa ligera"
Hazlo = "Llorar"
Quémalo = "Sí"
Nadando = "Sí"
Pepinillos = "Sí"


if (Carne == "Doble triple" and 
Salsa == "De Coral" and
Decorado == "De lujo" and
Estilo == "Animal" and 
Guijarros == "Extra" and
Vibración == "Exprimida" and
Eje == "Con grasa ligera" and 
Hazlo == "Llorar" and
Quémalo == "Sí" and
Nadando == "Sí" and
Pepinillos == "No"):
  print("Quiere la kangreburguer")
else:
  print("No quiere la kangrebuger")
```
