```
#Importamos paquetes
import matplotlib.pyplot as plt
#import math as mt
from math import *
```

```
# Funciones matematicas a graficar
#0.8sin(200 - 2πx)
#1.1cos(250 - πx)
```

```
X = []
Y_sin = []
Y_cos = []
x = -10
while(x <= 10):
  X.append(x)
  Y_sin.append( 0.8*sin(200 - 2*x*pi) )
  Y_cos.append( 1.1*cos(250 - pi*x) )
  x += 0.01
```

```
print(X)
print(Y_sin)
print(Y_cos)
```

```
plt.plot(X,Y_sin)
```

```
plt.plot(X,Y_cos)
```

```
plt.plot(X,Y_sin, label = 'y = 0.8sin(200 - 2πx')
plt.plot(X,Y_cos, label = 'y = 1.1cos(250 - πx)')
plt.legend()
```

```
Y_sup = []
for n in range(len(X)):
  Y_sup.append(Y_sin[n] + Y_cos[n])
print(Y_sup)
```

```
plt.plot(X, Y_sup, label = 'Pulsos de ondas')
plt.legend()
```
