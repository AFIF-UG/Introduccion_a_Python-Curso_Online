```
#Importamos paquetes
import numpy as np
import matplotlib.pyplot as plt
```

```
def abri_arc(nombre_archivo):
  f = open(nombre_archivo, 'r')
  return f
```

```
def datos_a_matriz(f):
  matriz = []
  matriz = [line.split() for line in f]
  return matriz
```

```
def cerrar_archivo(f):
  f.close()
```

```
def str_a_float(matriz):
  for n in range(len(matriz)):
    matriz[n][0] = float(matriz[n][0])
    matriz[n][1] = float(matriz[n][1])
```

```
def matriz_numpy(matriz):
  matriz = np.array(matriz)
  return matriz
```

```
def graficar(matriz):
  plt.plot(matriz[:,0], matriz[:,1], linewidth = 0.5)
  plt.scatter(matriz[:,0], matriz[:,1], c = 'r', s = 1)
```

```
f = abrir_arc('Archivo_datos.txt')
matriz = datos_a_matriz(f)
cerrar_archivo(f)
str_a_float(matriz)
matriz = matriz_numpy(matriz)
graficar(matriz)
```
